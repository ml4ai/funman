```python
import pandas as pd
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Set the size of the grid
size=10

# Create the parameter coordinates
p1 = pd.DataFrame(data={"p1" : np.linspace(0, 1, size+1)[0:-1]})
p2 = pd.DataFrame(data={"p2": np.linspace(0, 1, size+1)[0:-1]})

# Create a grid with the parameter coordinates
d1 = p1.merge(p2, how="cross")

# Set the value for each coordinate
d1['feasible'] =  [(1/(size) if bool(random.getrandbits(1)) else 0) for i in range(len(d1)) ]


# Plotting
fig, ax = plt.subplots()
ax.set_title("Feasible Parameter Space")
ax.set_xlabel("p1")
ax.set_ylabel("p2")
ax.set_xticks(p1['p1'])
ax.set_yticks(p2['p2'])
plt.gca().set_aspect('equal', adjustable='box')
ax.grid(True)
ax.scatter(d1['p1']+(1/((size)*2)), d1['p2']+(1/((size)*2)), s=7000*d1['feasible'], marker="s")
plt.show()
```
